```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
database_normalizada = pd.read_csv('resultado/dados_curitiba.csv', sep=";")
database_sem_normalizacao = pd.read_csv('resultado/dados_curitiba_sem_equalizacao.csv', sep=";")
```

```python
num_img = database_normalizada.Imagem.values.tolist()
diferenca_normalizado = database_normalizada.SIFT_SIFT_Diferenca.values.tolist()
database_nao_normalizacao = database_sem_normalizacao.SIFT_SIFT_Diferenca.values.tolist()

num_img = list(map(int, num_img))
```

```python
database = {
    'normalizado': diferenca_normalizado,
    'n√£o normalizado': database_nao_normalizacao
}
```

```python

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.stackplot(num_img, database.values(),
             labels=database.keys(), alpha=0.8)
ax.legend(loc='upper left')
ax.set_title('T√çTULO')
ax.set_xlabel('Imagens')
ax.set_ylabel('Diferen√ßa')

plt.show()
```

```python
sift_sift_inliers = database_normalizada.SIFT_SIFT_Inliers.values.tolist()
orb_orb_inliers = database_normalizada.ORB_ORB_Inliers.values.tolist()
akaze_akaze_inliers = database_normalizada.AKAZE_AKAZE_Inliers.values.tolist()
fast_brief_inliers  = database_normalizada.FAST_BRIEF_Inliers.values.tolist()
fast_sift_inliers  = database_normalizada.FAST_SIFT_Inliers.values.tolist()

sift_sift_tempo = database_normalizada.SIFT_SIFT_Tempo.values.tolist()
orb_orb_tempo = database_normalizada.ORB_ORB_Tempo.values.tolist()
akaze_akaze_tempo = database_normalizada.AKAZE_AKAZE_Tempo.values.tolist()
fast_brief_tempo  = database_normalizada.FAST_BRIEF_Tempo.values.tolist()
fast_sift_tempo  = database_normalizada.FAST_SIFT_Tempo.values.tolist()
```

```python
plt.plot(fast_sift_inliers , color='green', label='fast sift')
plt.plot(sift_sift_inliers, color='red', label='sift')
plt.plot(orb_orb_inliers, color='blue', label='orb')
plt.plot(akaze_akaze_inliers, color='orange', label='akaze')
plt.plot(fast_brief_inliers , color='purple', label='fast brief')
plt.xlabel('Inliers')
plt.ylabel('Valores')
plt.legend()
plt.show()
```

```python
def retorna_acertos_tempo(lista_inliers, lista_tempo):
    acertos = 0
    tempo_total = datetime.timedelta()
    for inlier, tempo in zip(lista_inliers, lista_tempo):
        tempo_obj = datetime.datetime.strptime(tempo, "%H:%M:%S.%f").time()
        delta = datetime.timedelta(
                hours=tempo_obj.hour,
                minutes=tempo_obj.minute,
                seconds=tempo_obj.second,
                microseconds=tempo_obj.microsecond
            )
        tempo_total += delta

        if inlier >= 20:
            acertos += 1
    return acertos, tempo_total

sift_sift_inliers_acertos, sift_sift_inliers_tempo = \
    retorna_acertos_tempo(sift_sift_inliers, sift_sift_tempo)

orb_orb_inliers_acertos, orb_orb_inliers_tempo = \
    retorna_acertos_tempo(orb_orb_inliers, orb_orb_tempo)

akaze_akaze_inliers_acertos, akaze_akaze_inliers_tempo = \
    retorna_acertos_tempo(akaze_akaze_inliers, akaze_akaze_tempo)

fast_brief_inliers_acertos, fast_brief_inliers_tempo = \
    retorna_acertos_tempo(fast_brief_inliers, fast_brief_tempo)

fast_sift_inliers_acertos, fast_sift_inliers_tempo  = \
    retorna_acertos_tempo(fast_sift_inliers, fast_sift_tempo)
```
